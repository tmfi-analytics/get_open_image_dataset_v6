# 訓練用の画像と教師ラベルをダウンロードする。今回は人体検出のためのデータセット。

```python
import fiftyone

# 訓練用のデータセットをロードする
train = fiftyone.zoo.load_zoo_dataset(
    'open-images-v6', 
    split='train',  # train用のデータセットとしてダウンロード
    label_types=['detections'], # ['detections', 'classifications', 'relationships', 'segmentations']のうち、どれかを選ぶ。今回は物体検出のため、'detections'
    classes=['Human body'], # 検出したい対象のクラスを指定。今回は人体のため、人体のため、'Human body'
    max_samples=100 # 取得したいサンプル数を指定
)

# ロードしたデータセットをファイルに出力する
train.export(
   export_dir='data/',  # データセットを出力するフォルダを指定
   split='train',  # train用のデータセットを出力する
   dataset_type = fiftyone.types.YOLOv5Dataset,  # 教師ラベルの出力形式を指定。今回はYOLOv5用のフォーマット
   classes=['Human body'] # 検出したい対象のクラスを指定。
)
```

```python
# バリデーション用のデータセットをロードする
val = fiftyone.zoo.load_zoo_dataset(
    'open-images-v6', 
    split='validation',  # バリデーション用のデータセットとしてダウンロード
    label_types=['detections'], # ['detections', 'classifications', 'relationships', 'segmentations']のうち、どれかを選ぶ。今回は物体検出のため、'detections'
    classes=['Human body'], # 検出したい対象のクラスを指定。今回は人体のため、人体のため、'Human body'
    max_samples=100 # 取得したいサンプル数を指定
)

# ロードしたデータセットをファイルに出力する
val.export(
   export_dir='data/',  # データセットを出力するフォルダを指定
   split='validation',  # バリデーション用のデータセットを出力する
   dataset_type = fiftyone.types.YOLOv5Dataset,  # 教師ラベルの出力形式を指定。今回はYOLOv5用のフォーマット
   classes=['Human body'] # 検出したい対象のクラスを指定。
)
```
